```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from random import randint
import matplotlib.pyplot as plt
```

```python
#for 10 pages
pages = [str(i) for i in range(1,11)]
#list for storing data
hotel_name = []
room_price = []
ratings = []
for page in pages:
    #url
    url = "https://www.goibibo.com/hotels/hotels-in-prayagraj-ct/?pg="+page
    headers = {    
        'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/79.0.3945.79 Chrome/79.0.3945.79 Safari/537.36"
        }
    #request to get data
    response = requests.request("GET", url,headers = headers)
    
    #parse the downloaded data
    data = BeautifulSoup(response.text, 'html.parser')
    #get the sections with specified class name
    sections = data.find_all('div',attrs = {'class' : 'width100 fl htlListSeo hotel-tile-srp-container hotel-tile-srp-container-template new-htl-design-tile-main-block'})
    for sec in sections :
        #get hotel name and room price
        name = sec.find('p')
        price = sec.find('li', attrs = {'class' : 'htl-tile-discount-prc'})
        rat = sec.find('div',attrs = {'style' : 'float: right; margin-right:8px; margin-top:3px;'})
        ratings.append(rat.text.replace('\n','')[-5:-2])
        hotel_name.append(name.text)
        room_price.append(price.text)
df = pd.DataFrame({'Hotel Name':hotel_name, 'Room Price' : room_price, "Rating" : ratings})
```
